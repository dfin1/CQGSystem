## Dataframe construction:

```python
df = pd.read_csv('../../Data/DF_MIMICS_ConstructedFascets_v2_6.csv')
```

```python
df
```

```python
df_tmp = []
for idx, item in df.iterrows():
    trueFascets = item['fascets'].split(' , ')
    #Add True Fascets 
    for fascet in trueFascets:
        for i in range(FascetRepetitionFactor):
            el = {
                'query' : item['query'],
                'clariQ' : item['target'],  
                'fascet' : fascet,
                'isRelated' : 1
            }
            df_tmp.append(el)
    
    #Add False Fascets
    print("false Fascets")
    fascetsArr = item['fascets_20'].split(' , ')
    for idx, fascet in enumerate(fascetsArr):
        for i in range(FascetRepetitionFactor):
            el = {
                'query' : item['query'],
                'clariQ' : item['target'],  
                'fascet' : fascet,
                'isRelated' : 0
            }
            df_tmp.append(el)
df_full = pd.DataFrame(df_tmp)
```

```python
df_full = df_full.sample(frac=1)
```
