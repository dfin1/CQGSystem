```python
import json
import pandas as pd
import requests
import random
import csv

```

```python
def callFascetGenerationAPI(body):
    response = requests.post("http://128.119.246.170:7789/extract", json = body )

    if response.status_code !=200:
        print("Error status code:" + str(response.status_code))
    
    #print(response)
    #return only the fasctes
    return (response.json())
```

## Dataframe loading:

```python
df = pd.read_csv("../DF_MIMICS_Full_constructed.csv")
#df = df[0:1000]
```

```python
df
```

## Fascet Generation:

```python
fascets_5 = []
#fascets_12 = []
#fascets_20 = []
clariq = []

for idx, item in df.iterrows():
    try:
        #print(item['source'].split(' | ')[0])
        body = { "query": item['query']}
        response = callFascetGenerationAPI(body)
        
        #print(response)
        #print(response)
        fascets_5.append(' , '.join(response['bestFascets'].split( " , ")[0:5]))
        #fascets_12.append(' , '.join(response['fascets'].split(" , ")[0:12]))
        #fascets_20.append(response['fascets'])
        clariq.append(response['clariQ'])
        print(response['query'] + " => " + response['clariQ'][0])
        
    except:
        print(idx)
        fascets_5.append(" ")
        #fascets_12.append(" ")
        #fascets_20.append(" ")
        clariq.append(" ")

#print(fascets_5)
#print(len(fascets_5))
try:        
    df['fascets_5'] = fascets_5
    #df['fascets_12'] = fascets_12
    #df['fascets_20'] = fascets_20
    df['clariq'] = clariq
except:
    df_f20 = pd.DataFrame(fascets_5)
    df_f20.to_csv("DF_MIMICS_constructed_FinPipeline_f5.csv")
    df_clq = pd.DataFrame(clariq)
    df_clq.to_csv("DF_MIMICS_constructed_FinPipeline_clq.csv")
```

```python
df.to_csv("DF_MIMICS_constructed_FinPipeline.csv")
```
